```python
# Import the robot class we are developing
import robot

# Make a new leg variable which is a robot.Leg class
leg = robot.Leg()
```

```python
a=3.1415/180
leg.set_joint_pos(90*a, 90*a,0,0,0,0)
```

```python
#this can accumulate the spring energy
import time
time.sleep(3)

leg.set_joint_pos(20*a,160*a,0,0,-5,5)
time.sleep(0.8)
leg.set_joint_pos(80*a,100*a,0,0,5,-5)
time.sleep(0.15)   #This time sleep will help the spring accumulate the energy step by step

leg.set_joint_pos(20*a,160*a,0,0,-5,5)
time.sleep(1)
leg.set_joint_pos(80*a,100*a,0,0,5,-5)
time.sleep(0.2)

leg.set_joint_pos(20*a,160*a,0,0,-5,5)
time.sleep(0.8)
leg.set_joint_pos(80*a,100*a,0,0,5,-5)
time.sleep(0.23)

leg.set_joint_pos(20*a,160*a,0,0,-5,5)
time.sleep(0.8)
leg.set_joint_pos(80*a,100*a,0,0,5,-5)
time.sleep(0.23)

leg.set_joint_pos(20*a,160*a,0,0,-5,5)
time.sleep(0.8)
leg.set_joint_pos(80*a,100*a,0,0,5,-5)
time.sleep(0.3)
```
